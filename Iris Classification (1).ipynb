```python
# Step 1 : Import Libraries

import numpy as np
import pandas as pd

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder,StandardScaler

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
```

```python
# Step 2 :Load Dataset

iris = load_iris()

x = iris.data
y = iris.target

# Create DataFrame 
df = pd.DataFrame(x,columns = iris.feature_names)

# Add target column

df['Target'] = y

# Display first Five rows
df.head()

```

```python
# Step 3 : Preprocessing

# Check null values 
print(df.isnull().sum())

# Check Duplicate values
print(df.duplicated().sum())

# Dataset Shape 
print(df.shape)
# Dataset Info
print(df.info())
```

```python
# Step 4 : Separate input and output

x = df.drop('Target',axis = 1)
y = df['Target']
```

```python
# Step 5 : Label Encoding

le = LabelEncoder()

y = le.fit_transform(y)

# One-hot-encoding

y_encoded = to_categorical(y)
print(y_encoded[:5])
```

```python
# Step 6 : Normalize features
scaler = StandardScaler()

X_scaled = scaler.fit_transform(x)
print(X_scaled[:5])
```

```python
# Step 7 :Train-test split

X_train,X_test, y_train, y_test = train_test_split(
    X_scaled,
    y_encoded,
    test_size = 0.2,
    random_state = 42
)

print(X_train.shape)
print(X_test.shape)
```

```python
# Step 8: Build the Model

model = Sequential()
# Hidden Layer 1 
model.add(Dense(10,input_shape=(4,),activation = 'relu'))

# Hidden Layer 2 
model.add(Dense(8,activation='relu'))

# Output Layer 
model.add(Dense(3,activation = 'softmax'))
```

```python
# Step 9 : Compile The model 

model.compile(
    optimizer = 'adam',
    loss = 'categorical_crossentropy',
    metrics = ['accuracy']
)
```

```python
# Step 10 : Train the model

history = model.fit(
    X_train,
    y_train,
    epochs = 50,
    batch_size =8,
    validation_split = 0.2
)
```

```python
# Step : 11 Evaluate the model

loss,accuracy = model.evaluate(X_test, y_test)

print("Loss:",loss)
print("Accuracy",accuracy)
```

```python
# Step 12: Prediction

y_pred = model.predict(X_test)

# Convert predicted values
y_pred_classes = np.argmax(y_pred, axis = 1)

# Convert avtual values
y_test_classes = np.argmax(y_test, axis = 1)

print("Predicted Values:")
print(y_pred_classes[:10])

print("Actual Values :")
print(y_test_classes[:10])
```

```python
# Step 13 : Accuracy Score

from sklearn.metrics import accuracy_score

acc = accuracy_score(y_test_classes,y_pred_classes)
print("Accuracy Score:",acc)
```

```python
# Step 14 : Classification Report

from sklearn.metrics import classification_report

print(classification_report(y_test_classes,y_pred_classes))
```

```python
import joblib
from tensorflow.keras.models import load_model

# Save model
model.save("iris_ann_model.h5")

# Save scaler
joblib.dump(scaler, "iris_scaler.pkl")

print("Model and scaler saved successfully!")


loaded_model = load_model("iris_ann_model.h5")
loaded_scaler = joblib.load("iris_scaler.pkl")


unseen_flower = np.array([[6.1, 2.8, 4.7, 1.2]])

# Scale the input
unseen_flower_scaled = loaded_scaler.transform(unseen_flower)

# Predict
prediction = loaded_model.predict(unseen_flower_scaled)

predicted_class = np.argmax(prediction, axis=1)[0]

# Convert class index to species name
species_names = ['setosa', 'versicolor', 'virginica']
predicted_species = species_names[predicted_class]

print("\nUnseen Flower Sample:")
print(unseen_flower)

print("\nPrediction Probabilities:")
print(prediction)


print("\nPredicted Species:", predicted_species)
```
